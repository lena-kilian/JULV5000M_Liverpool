Import libraries

```python
import pandas as pd
import numpy as np
import requests
import json
import os
import geopandas
```

```python
# create 'data/' folder 
def createFolder(directory):
    try:
        if not os.path.exists(directory):
            os.makedirs(directory)
    except OSError:
        print ('Error: Creating directory. ' +  directory)

createFolder('./data/')
```

```python
# get data and write to files

rs_url = 'http://opendatacommunities.org/downloads/cube-table?uri=http%3A%2F%2Fopendatacommunities.org%2Fdata%2Fhomelessness%2Frough-sleeping%2Fcount'
ic_url = 'http://opendatacommunities.org/downloads/cube-table?uri=http%3A%2F%2Fopendatacommunities.org%2Fdata%2Fsocietal-wellbeing%2Fdeprivation%2Fimd-income-score-2010'

for url in [rs_url, ic_url]:
    # request object
    r = requests.get(url)
    # pull file name
    name = url.split('%2F')[-2] + '_' + url.split('%2F')[-1]
    # Open target file for writing binary data
    fo = open('data/' + name + '.csv', 'wb')
    # Write file content
    fo.write(r.content)
    # Close file
    fo.close()
```
